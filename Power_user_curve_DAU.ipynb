```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import math
from matplotlib.ticker import PercentFormatter
from datetime import datetime,timedelta
from matplotlib import dates as dt
```

```python
fp=pd.read_csv(r"C:\Users\Jarvis\Desktop\power.csv",names=[i for i in range(1,32)],skiprows=1,engine='python')
fp.head()
```

```python
#Total number of user for each day
t=fp.apply(pd.value_counts).shape
Total=t[0]
print(Total)
fp.shape
```

```python
#Number of user for each day
length_of_column=[]
for j in range(1,32):
    t=fp[j]
    s1=[i for i in fp[j] if i==i]
    length_of_column.append(len(s1))
```

```python
#Percentage of user for each day
percentage=[]
for i in range(31):
    percentage.append((length_of_column[i]/Total)*100)
    

```

```python
%matplotlib inline
sns.set()
d=[i for i in range(1,32)]
plt.figure(figsize=(20,3))
plt.bar(d,length_of_column,width=0.8,align='center',color='red',alpha=0.25)
plt.xticks([r + 0.9 for r in range(len(d))],d,rotation=24)
plt.yticks([i for i in range(2000,11000,1000)])
plt.ylabel("Number of users in each day")
plt.xlabel('Total active days in a month')
plt.subplots_adjust(bottom=0.0000009,top=1.99)
plt.plot(d,length_of_column,color='black')
plt.title("Power user curve")
for i,j in zip(d,length_of_column):
    plt.annotate(str(j),xy=(i,j),color='red')
plt.savefig("power_user_curve.png",bbox_inches='tight')
```
